#### Q1.1 Age-Normalized Risk Index (ANRI) (5 points)

The **Age-Normalized Risk Index (ANRI)** identifies patients whose stroke risk is unusually high for their age.  
A higher ANRI indicates a higher stroke risk relative to age.

*Hint: ANRI is calculated by dividing each patient’s `stroke_risk_pct` by their `age`.*

**Tasks:**
1. Compute a new column `ANRI` (handle divide-by-zero safely).  
2. Plot the distribution of ANRI using a **histogram** (with KDE).   
3. Show the **top 10 patients** with the highest ANRI values and briefly interpret the results.

```python
def compute_anri(df):
    """
    Compute and visualize the Age-Normalized Risk Index (ANRI)
    """
    
    anri = df['stroke_risk_pct']/df['age']
    df['anri'] = anri

    anri_top10 = df.sort_values(by='anri', ascending=False)
    print('--- Top 10 with the highest ANRI values ---')
    print(anri_top10.head(10)[['age','anri']],'\n')

    print(anri_top10['anri'].describe())
    
    sns.histplot(
        df['anri'],
        kde= True,
        bins=30,
        color='navy'
    )
    plt.xlabel('ANRI')
    plt.title('Distribution of ANRI')
    plt.show()
    
    return df

df = compute_anri(df)
```

### _***Interpretation***_
We observe that patients with the top 10 ANRI values are mostly in their 30s, with ANRI scores around 3. From the overall ANRI distribution, these ten patients are outside three standard deviations from the mean and can be considered outliers. The overall mean ANRI is 1.41, with approximately 68% of patients' ANRI values within the range of [1.00, 1.82]. This suggests that around 68% of patients have a stroke risk percentage roughly equal to 1.0–1.8 times their age.

#### Q1.2 Chronic Condition Score (CCS) (5 points)

The **Chronic Condition Score (CCS)** represents the number of chronic cardiovascular conditions a patient has, specifically high blood pressure and irregular heartbeat.  
A higher CCS indicates greater chronic disease burden.

*Hint: CCS is the sum of the binary columns `high_bp` and `irregular_heartbeat`.*

**Tasks:**
1. Add a new column `CCS` to the DataFrame.
2. Compute the **average stroke risk percentage** for each CCS level (0, 1, 2).  
3. Visualize the results using a **bar plot**, and interpret how stroke risk changes with higher CCS.

```python
def compute_ccs(df):
    """
    Compute and visualize the Chronic Condition Score (CCS)
    """
    ccs = df['high_bp'] + df['irregular_heartbeat']
    df['ccs'] = ccs
    ccs_summary = df.groupby('ccs')[['age','stroke_risk_pct']].mean()

    title_prefix = 'Average Stroke Risk Percentage'
    print(f'--- {title_prefix} for Each Level ---')
    print(ccs_summary.sort_values(by='stroke_risk_pct',ascending=False))
    
    sns.barplot(
        x = ccs_summary.index,
        y = ccs_summary['stroke_risk_pct'].values,
        palette='Blues'
    )
    
    plt.title(f'{title_prefix} Bar Plot (CCS)')
    plt.xlabel('level')
    plt.ylabel('Percentage (%)')
    plt.tight_layout()
    plt.show()
    
    return ccs_summary

ccs_summary = compute_ccs(df)
```

### _***Interpretation***_
We observed a clear positive relationship between CCS values and the average stroke risk percentage. The CCS metric is derived from the sum of two features — high_bp(high blood pressure) and irregular heartbeat. Therefore, it is reasonable to infer that patients with both conditions have an average stroke risk of about 86%. Those with only one of the two conditions still face a relatively high average risk of around 72%, while even patients without either condition maintain an average stroke risk of roughly 50%, possibly due to the influence of other underlying health factors.

#### Q1.3 Symptom Burden Index (SBI) (5 points)

The **Symptom Burden Index (SBI)** quantifies the total number of symptoms reported by each patient.  
A higher SBI indicates a greater overall symptom load, which is expected to relate to higher stroke risk.

*Hint: SBI is the sum of all symptom indicator columns (values 0 or 1).*

**Tasks:**
1. Compute the total `SBI` for each patient using all symptom columns. Categorize patients into three groups based on SBI - *Low (0–3)*; *Moderate (4–6)*; *High (7+)*
2. Analyze how the **average stroke risk percentage** changes across these SBI groups 
3. Compute the **correlation** between `SBI` and `stroke_risk_pct`. 

```python
def compute_sbi(df):
    """
    Compute the Symptom Burden Index (SBI)
    """

    df_sbi = df.copy()
    df_sbi = df_sbi.drop('at_risk',axis=1)
    indicator_cols = [col for col in df_sbi.columns if set(df_sbi[col].dropna().unique()).issubset({0,1})]
    sbi = df_sbi[indicator_cols].sum(axis=1)
    df['sbi'] = sbi

    for i in range(df.shape[0]):
        if df.loc[i,'sbi'] <= 3:
            df_sbi.loc[i, 'sbi_group'] = 'Low'
        elif df.loc[i,'sbi'] >=4 and df.loc[i,'sbi'] <=6:
            df_sbi.loc[i, 'sbi_group'] = 'Moderate'
        else:
            df_sbi.loc[i, 'sbi_group'] = 'High'

    sbi_summary = df_sbi.groupby('sbi_group')[['age','stroke_risk_pct']].mean().sort_values(by='stroke_risk_pct',ascending = True)

    corr_matrix = np.corrcoef(df['sbi'],df['stroke_risk_pct'])

    title_prefix = 'Average Stroke Risk Percentage'
    print(f'--- {title_prefix} Changes across SBI Groups ---')
    print(sbi_summary.sort_values(by='stroke_risk_pct',ascending = False))
    
    print(f'\nCorrelation b/w SBI & stroke risk pct: {corr_matrix[0][1]:.4f}')

    sns.barplot(
        x = sbi_summary.index,
        y = sbi_summary['stroke_risk_pct'].values,
        palette='Blues'
    )
    
    plt.title(f'{title_prefix} Bar Plot (SBI)')
    plt.xlabel('level')
    plt.ylabel('Percentage (%)')
    plt.tight_layout()
    plt.show()
    
    return sbi_summary

sbi_summary = compute_sbi(df)
```

### _***Interpretation***_
We observed that the likelihood of stroke increases sharply as the number of health-related symptoms (features) rises. Patients classified as High Risk, with more than seven symptom indicators, have an incidence rate of 99%, nearly 100%. This suggests that patients in this group should pay highly close attention to their health status. Even those individuals in the Moderate level group face a substantial stroke probability of about 81%, while the Low Risk group still exhibits an average risk of roughly 50%.  

Furthermore, the correlation between the Symptom Burden Index (SBI) and stroke risk percentage reaches 0.68, indicating a strong positive relationship between these two variables.

#### Q1.4 Symptom Predictive Power using Mutual Information (MI) (5 points)

**Mutual Information (MI)** measures how informative a feature is for predicting a target variable.  
Here, it shows how strongly each symptom relates to the stroke risk label (`at_risk`).

*Hint: Use *`mutual_info_classif`* from *`sklearn.feature_selection`* to compute MI scores.*

**Tasks:**
1. Compute the 'MI' score between each symptom and `at_risk`.  
2. Sort symptoms by MI score and list the **Top 10** most predictive ones.  

```python
def compute_mi(df):
    """
    Compute the Mutual Information (MI) score
    """
    
    from sklearn.feature_selection import mutual_info_classif
    target = df['at_risk']
    df_mi = df.iloc[:,:18]
    
    indicator_cols = [col for col in df_mi.columns if set(df_mi[col].dropna().unique()).issubset({0,1})]
    mi_scores = mutual_info_classif(df_mi[indicator_cols], target, discrete_features= True)
    mi_summary = pd.DataFrame({
            'Features': indicator_cols,
            'MI_Scores': mi_scores
            }).sort_values(by='MI_Scores', ascending=False)

    print('--- Top 10 MI Scores ---')
    print(mi_summary.head(10))
    
    return mi_summary

mi_summary = compute_mi(df)
```

#### Q1.5 Age-Adjusted Risk Z-Scores (AARZ) (5 points)

The **Age-Adjusted Risk Z-Score (AARZ)** compares a patient’s stroke risk to others in the same age group.  
It highlights patients whose risk levels are unusually high or low relative to their peers.

*Hint: For each 10-year age group, calculate how many standard deviations a patient’s stroke risk percentage is above or below the group’s mean value.*

**Tasks:**
1. Create 10-year age bins, within each age group, compute the *Z-score* for stroke risk (`stroke_risk_pct`).  
2. Identify the **Top 5 patients** with the highest Z-scores (highest relative risk).  
3. For **each age group**, list the **Top 2 patients** with the highest Z-scores. 

```python
print(min(df['age']),max(df['age']),sep=' , ')
```

```python
def compute_aarz(df):
    """
    Compute Age-Adjusted Risk Z-Scores (AARZ)
    """
    df_age = df.copy()
    df_age['age_group'] = pd.cut(df_age['age'], bins=range(10, 100, 10))
    stats = df_age.groupby('age_group')['stroke_risk_pct'].agg(['mean','std']).reset_index()
    df_age = pd.merge(df_age,stats)
    df_age['aarz'] = (df_age['stroke_risk_pct'] - df_age['mean']) / df_age['std']
    df_age['aarz'] = np.where(df_age['std']>0,  (df_age['stroke_risk_pct'] - df_age['mean']) / df_age['std'], 0.0)
    
    df['aarz'] = df_age['aarz']

    top5_outliers = df_age.sort_values(by='aarz',ascending = False).head(5)
    top_by_group = (
        df_age.sort_values(by='aarz',ascending=False)
                .groupby('age_group')
                .head(2)
                .sort_values(by='age_group')
    )
    display(top5_outliers[['aarz','age_group']])
    display(top_by_group[['aarz','age_group']])

    return top5_outliers,top_by_group

top5_outliers, top_by_group = compute_aarz(df)
```

#### Q1.6 Risk Consistency Index (RCI) (5 points)

The **Risk Consistency Index (RCI)** measures how well the continuous stroke risk scores align with the binary label `at_risk`.  
It captures how distinctly the two groups: *at risk* and *not at risk*, differ in their average stroke risk percentage.
Higher RCI values indicate stronger consistency between risk scores and the `at_risk` label.

*Hint: RCI is calculated by taking the absolute difference between the group means and dividing it by the pooled standard deviation.*  

**Tasks:** 
1. Compute the mean and standard deviation of `stroke_risk_pct` for both groups and 
2. Calculate the `RCI`.
2. Interpret whether stroke risk percentages are consistent with the binary classification labels.

```python
def compute_rci(df):
    """
    Compute the Risk Consistency Index (RCI) 
    """

    stats = df.groupby('at_risk')['stroke_risk_pct'].agg(['mean','std', 'count'])
    m0, s0, n0 = stats.loc[0, ['mean','std', 'count']]
    m1, s1, n1 = stats.loc[1, ['mean','std', 'count']]
    pooled_std = np.sqrt(((n0-1)* s0**2 + (n1-1)* s1**2) / (n0 + n1 -2))
    RCI = np.abs(m1-m0) / pooled_std

    print(f"RCI value:{RCI:.3f}")
    
    return RCI

RCI = compute_rci(df)
```

### _***Interpretation***_
The RCI value is a relatively high value 2.87, indicating a significant difference in stroke risk percentages between the two groups (at_risk = 0 and at_risk = 1). The distributions of these two groups are nearly not overlapped, suggesting that the stroke_risk_pct can almost perfectly distinguish whether a patient belongs to which at_risk category is.

#### Q1.7 Composite Health Risk Index (CHRI) (5 points)

The **Composite Health Risk Index (CHRI)** combines multiple cardiovascular indicators into a single, weighted risk score.  
It integrates *high blood pressure*, *irregular heartbeat*, and *age-normalized risk (ANRI)* to capture an individual’s overall health vulnerability.

*Hint:* Compute CHRI using the weighted formula:  *CHRI = 0.4 * High Blood Pressure + 0.4 * Irregular Heartbeat + 0.2 * ANRI*

**Tasks:**
1. Calculate the **Composite Health Risk Index (CHRI)** for each patient.   
2. Find the **Top 5 patients** with the highest CHRI values.
3.   Compute correlations between `CHRI` and both `stroke_risk_pct` and `at_risk`.

```python
def compute_chri(df):
    """
    Compute the Composite Health Risk Index (CHRI) 
    """
    df_new = df.copy()
    chri = 0.4*df_new['high_bp'] + 0.4*df_new['irregular_heartbeat'] + 0.2*df_new['anri']
    df_new['chri'] = chri
    top_chri = df_new.sort_values(by ='chri', ascending= False).head(5)
    corr_chri_risk = np.corrcoef(df_new['chri'], df_new['at_risk'])
    corr_chri_stroke = np.corrcoef(df_new['chri'], df_new['stroke_risk_pct'])

    print(f"CHRI & at_risk : {corr_chri_risk[0][1]:.4f}")
    print(f"CHRI & stroke_risk_pct : {corr_chri_stroke[0][1]:.4f}")
    
    return df_new, top_chri, corr_chri_risk, corr_chri_stroke
    
df_new, top_chri, corr_chri_risk, corr_chri_label = compute_chri(df)
```

#### Q1.8 Risk Stratification Buckets (5 points)

The **Risk Stratification Buckets** divide patients into *Low*, *Moderate*, and *High* risk groups based on their normalized stroke risk percentage.  
This helps compare how *age*, *symptom burden (SBI)*, and *chronic conditions (CCS)* vary across different risk levels.

*Hint: Normalize `stroke_risk_pct` by dividing by 100, then assign each patient to a risk category.*

**Tasks:**
1. Categorize patients into three buckets — *Low (0–0.3)*, *Moderate (0.3–0.7)*, *High (0.7–1.0)*. For each group, compute the average **age**, **SBI**, and **CCS**.  
2. Visualize these averages using a bar plot with a log scale on the y-axis and interpret how patient characteristics differ by risk level.

```python
def compute_risk_buckets(df):
    """
    Create risk stratification buckets based on normalized stroke risk percentage
    """
    df_new = df.copy()
    df_new['stroke_risk_pct_normalized'] = df_new['stroke_risk_pct'] / 100

    for i in range(df.shape[0]):
        if df_new.loc[i,'stroke_risk_pct_normalized'] <= 0.300:
            df_new.loc[i, 'pct_normalized_group'] = 'Low'
        elif df_new.loc[i,'stroke_risk_pct_normalized'] >=0.301 and df_new.loc[i,'stroke_risk_pct_normalized'] <=0.700:
            df_new.loc[i, 'pct_normalized_group'] = 'Moderate'
        else:
            df_new.loc[i, 'pct_normalized_group'] = 'High'

    bucket_summary = df_new.groupby('pct_normalized_group')[['age','sbi','ccs']].mean().sort_values(by='age',ascending = True)
  
    return bucket_summary

bucket_summary = compute_risk_buckets(df)
```

#### Q2.2 Optimal Number of Clusters (Elbow & Silhouette Methods) (20 points)

To determine the appropriate number of clusters (**k**), we use two evaluation methods:

- **Elbow Method:** observes the point where inertia (within-cluster variance) stops decreasing sharply.  
- **Silhouette Score:** measures how well clusters are separated (higher = better).  

**Tasks:**
1. Run **K-Means** clustering for values of \(k = 2\) to \(10\).  
2. Compute and store **Inertia** and **Silhouette Score** for each k.  
3. Visualize both metrics to identify the k that balances compactness and separation.  
4. Display the numeric summary table.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

def evaluate_kmeans_clusters(X_scaled, k_min=2, k_max=10):
    """
    Evaluate optimal number of clusters (k) using Elbow and Silhouette methods.
    Computes inertia and silhouette scores for k in [k_min, k_max].
    """
    k_values = range(k_min, k_max + 1)
    inertias = []
    silhouette_scores = []

    for k in k_values:
        km = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)
        labels = km.fit_predict(X_scaled)
        inertias.append(km.inertia_)
        
        score = silhouette_score(X_scaled, labels, metric='euclidean')
        silhouette_scores.append(score)
    
    opt_df = pd.DataFrame({
        "k": k_values,
        "inertia": inertias,
        "silhouette": silhouette_scores
    })

    display(opt_df.style.set_caption('Numeric Summary Table'))

    fig, axes = plt.subplots(2,1,figsize=(10, 12))
    axes[0].plot(k_values, silhouette_scores, 'bo-')
    axes[0].set_xlabel('Number of Clusters')
    axes[0].set_ylabel('Silhouette Score')
    axes[0].set_title('Silhouette Score vs Number of Clusters')
    axes[0].grid(True)
    

    axes[1].plot(k_values, inertias, 'bo-')
    axes[1].set_xlabel('Number of Clusters')
    axes[1].set_ylabel('Inertia')
    axes[1].set_title('Inertia vs Number of Clusters')
    axes[1].grid(True)

    plt.tight_layout()
    plt.show()
    
    return opt_df

opt_df = evaluate_kmeans_clusters(X_scaled)
```

#### Q2.3 K-Means Clustering and Visualization (15 points)

Using the results from the previous analysis, choose an appropriate number of clusters (**k**, typically 3–5).

**K-Means** partitions patients into groups such that individuals in the same cluster are more similar to each other based on selected numeric features, while those in different clusters are more dissimilar.

**Tasks:**
1. Fit **K-Means** on the standardized feature matrix (`X_scaled`) with the chosen value of **k**.  
2. Reduce the dataset to 2 dimensions using **PCA** for visualization.  
3. Assign cluster labels to each patient and map descriptive names (e.g., *Low Risk*, *Moderate Risk*, *High Risk*).  
4. Plot the clusters in 2D PCA space and highlight the cluster centroids.

```python
from sklearn.decomposition import PCA
def perform_kmeans_clustering(df, X_scaled, k=3):
    """
    Fit K-Means clustering, project to 2D PCA space, and visualize clusters.
    Returns DataFrame with cluster labels and centroids in PCA space.
    """
    
    km = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)
    km_labels = km.fit_predict(X_scaled)
    km_centers = km.cluster_centers_

    pca = PCA(n_components=2, random_state=42)
    X_pca = pca.fit_transform(X_scaled)
    expl_var = pca.explained_variance_ratio_.sum()
    centers_pca = pca.transform(km_centers)

    center_norms = np.linalg.norm(km_centers, axis=1)
    order = np.argsort(center_norms) #sort the center norm, the higher risk is, the higher its center norm value is

    results_df = pd.DataFrame({
    'pc1': X_pca[:, 0],
    'pc2': X_pca[:, 1],
    'cluster': km_labels
    })

    mapping = {
        2: 'Low Risk',
        1: 'High Risk',
        0: 'Moderate Risk', 
        3: 'Very High Risk'
    }

    results_df['risk_label'] = results_df['cluster'].map(mapping)
        
    df['cluster'] = results_df['cluster']
    df['risk_label'] = results_df['risk_label']

    print(f"PC1+PC2 explain {expl_var:.1%}")
    
    for c in range(k):
        mask = (results_df['cluster'] == c)
        plt.scatter(
            results_df.loc[mask, 'pc1'],
            results_df.loc[mask, 'pc2'],
            s=25,
            alpha=0.5,
            label=f"Cluster {c} – {mapping[c]}"
        )


    plt.scatter(
        centers_pca[:, 0],
        centers_pca[:, 1],
        s=100,
        marker='X',
        edgecolor='black',
        linewidth=1.0,
        label="Centroids"
    )
    
    plt.title(f'K-Means (k={k}) on standardized features\nVisualized by PCA 2D',fontsize=14)
    plt.xlabel('PC1',fontsize=13)
    plt.ylabel('PC2',fontsize=13)
    plt.xticks(fontsize=11)
    plt.yticks(fontsize=11)
    plt.grid(True, alpha=0.3)
    plt.legend(loc='best', fontsize=8,title_fontsize=8)
    plt.tight_layout()
    plt.show()


    return df, centers_pca

df, centroids_pca = perform_kmeans_clustering(df, X_scaled, k=4)
```

#### Q2.6 Gaussian Mixture Model (GMM) Comparison (25 points)

While K-Means assigns each point to exactly one cluster,  
a **Gaussian Mixture Model (GMM)** allows *probabilistic* membership, capturing overlap between patient groups.

**Tasks:**
1. Fit a GMM with the same number of clusters (**k**) used in K-Means.  
2. Align GMM cluster labels with K-Means labels for consistency.  
3. Compute the **Adjusted Rand Index (ARI)** to quantify agreement between the two models.  
4. Visualize both K-Means and GMM results side by side in PCA space.

```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics.cluster import contingency_matrix
from scipy.optimize import linear_sum_assignment

def compare_kmeans_gmm(df, X_scaled, k=3):
    """
    Compare K-Means and Gaussian Mixture Model (GMM) clustering.
    Computes Adjusted Rand Index (ARI) and visualizes both cluster assignments.
    """

    gmm_model = GaussianMixture(n_components=k, covariance_type='full', init_params='kmeans', random_state=42)

    gmm_labels = gmm_model.fit_predict(X_scaled)

    kmeans_model = KMeans(init = 'k-means++', n_clusters = k , n_init = 'auto',random_state=42)
    kmeans_labels = kmeans_model.fit_predict(X_scaled)
    
    ari_aligned = adjusted_rand_score(gmm_labels, kmeans_labels)

    c = contingency_matrix(gmm_labels, kmeans_labels) # 4x4 matrix
    row_ind, col_ind = linear_sum_assignment(-c)       
    
    mapping = {gmm_lab: km_lab for km_lab, gmm_lab in zip(row_ind, col_ind)}
    gmm_labels_aligned = np.vectorize(mapping.get)(gmm_labels)
    

    df['kmean_labels'] = kmeans_labels
    df['gmm_labels'] = gmm_labels
    df['gmm_labels_aligned'] = gmm_labels_aligned

    pca = PCA(n_components=2, random_state=42)
    X_pca = pca.fit_transform(X_scaled)

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    sns.scatterplot(
        x = X_pca[:,0],y = X_pca[:,1],
        hue = kmeans_labels,
        palette = 'icefire',
        s = 25,
        legend = 'full',
        ax = axes[0]
    )
    axes[0].set_xlabel('PCA-1')
    axes[0].set_ylabel('PCA-2')
    axes[0].legend(fontsize=12, loc='lower left')
    axes[0].set_title(f'K-Means (k={k})')


    sns.scatterplot(
        x = X_pca[:, 0], y = X_pca[:, 1],
        hue = gmm_labels_aligned,
        palette='icefire',
        s = 25,
        legend = 'full',
        ax = axes[1]
    )

    axes[1].set_xlabel('PCA-1')
    axes[1].set_ylabel('PCA-2')
    axes[1].legend(fontsize=12, loc='lower left')
    axes[1].set_title(f'GMM (k={k})')

    fig.suptitle(f"K-Means and GMM Visualization | ARI = {ari_aligned:.3f}", fontsize=16)
    plt.tight_layout()
    plt.show()
    
    return df ,ari_aligned


df, ari_aligned = compare_kmeans_gmm(df, X_scaled, k=4)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

def prepare_symptom_features(df):
    """
    Select and scale symptom-related features for modeling.
    """

    indicator_cols = [col for col in df.columns if set(df[col].dropna().unique()).issubset({0,1})]
    indicator_cols = [c for c in indicator_cols if c != 'at_risk']
    scaler = StandardScaler()
    preprocessor = Pipeline(steps=[('scaler', scaler)])
    display(indicator_cols)
    
    return indicator_cols, preprocessor

symptom_cols, preprocessor = prepare_symptom_features(df)
```

#### Q3.2 Regression – Predict Stroke Risk Percentage (20 points + bonus points)

Train regression models using symptom-only features to predict the continuous variable `stroke_risk_pct`.  
Evaluate them using **RMSE**, **MAE**, and **R²**.

At the minimum use the following models:
* k-nearest neighbors
* random forest
* linear regression

5 bonus points for each additional model you use, for up to 2 additional models.

Of course you are free to explore even more models.

```python
!pip install neural_network
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def train_regressors(df, symptom_cols, preprocessor):
    """
    Train and evaluate multiple regression models for stroke risk prediction.
    """
    
    X = df[symptom_cols]
    targets = df['stroke_risk_pct']

    X_train, X_test, y_train, y_test = train_test_split(
    X, targets, test_size=0.2, random_state=42
    )
    
    models = {
    "Linear Regression": LinearRegression(),
    "KNN Regressor": KNeighborsRegressor(n_neighbors=5),
    "Random Forest": RandomForestRegressor(n_estimators=200, random_state=42),
    
    #Bonus models
    "MLP Regressor": MLPRegressor(
            loss="squared_error",
            hidden_layer_sizes=(100,32),
            activation="relu",
            max_iter=200,
            n_iter_no_change=5,
            early_stopping=True,
            random_state=42
            ),
    "Gradient Boosting": GradientBoostingRegressor(
            n_estimators=200,
            learning_rate=0.02,
            max_depth=5,
            validation_fraction= 0.1,
            n_iter_no_change = 5,
            random_state=42)
    }

    results = []
    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        mae = mean_absolute_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)

        results.append({
            "Model": name,
            "RMSE": round(rmse, 3),
            "MAE": round(mae, 3),
            "R²": round(r2, 3)
        })

    
    results_df = pd.DataFrame(results).sort_values(by="R²",ascending=False)
    display(results_df)
    return results_df

reg_df = train_regressors(df, symptom_cols, preprocessor)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.inspection import permutation_importance

def analyze_best_regressor(df, reg_df, symptom_cols, preprocessor):
    """
    Display top features for the best-performing regression model.
    """

    best_reg_name = reg_df['Model'][0]
    
    X = df[symptom_cols]
    targets = df['stroke_risk_pct']

    X_train, X_test, y_train, y_test = train_test_split(
    X, targets, test_size=0.2, random_state=42
    )

    model = MLPRegressor(
            loss="squared_error",
            hidden_layer_sizes=(100,32),
            activation="relu",
            max_iter=1000,
            early_stopping=True,
            random_state=42
            )
    model = model.fit(X_train,y_train)

    result = permutation_importance(model, X_test, y_test, n_repeats=10, random_state=42)
    
    df_features = pd.DataFrame({
        "symptom_features": symptom_cols,
        "importance": result.importances_mean
    }).sort_values(by="importance", ascending=False)

    print(f"Best Regression Name: {best_reg_name}")
    display(df_features)


    
    plt.figure(figsize=(8,6))
    sns.barplot(
        data=df_features,
        x="importance",
        y="symptom_features",
        palette="crest"
    )
    plt.title("Most Important Symptom Features (Permutation Importance - MLP)",fontsize=14)
    plt.xlabel("Mean Importance (Decrease in R²)")
    plt.ylabel("Feature")
    plt.yticks(fontsize=12)
    plt.tight_layout()
    plt.show()

    return best_reg_name

best_reg_name = analyze_best_regressor(df, reg_df, symptom_cols, preprocessor)
```

#### Q3.4 Classification – Predict Stroke Risk Category (20 points + bonus points)

Using the same symptom-only features, train classification models to predict whether a patient is **at risk of stroke (1)** or **not at risk (0)**.  
Evaluate using **AUC**, **Accuracy**, **F1**, and **Balanced Accuracy**.

Use the following models:
* Random Forest
* k-nearest neighbors
* logistic regression

5 bonus points for each additional model you use, for up to 2 additional models.

Of course you are free to explore even more models.

```python
!pip install xgboost
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import roc_auc_score, accuracy_score, f1_score, balanced_accuracy_score

def train_classifiers(df, symptom_cols, preprocessor):
    """
    Train and evaluate multiple classification models to predict at-risk status.
    """

    X = df[symptom_cols]
    targets = df["at_risk"]

    X_train, X_test, y_train, y_test = train_test_split(
    X, targets, test_size=0.2, random_state=42
    )

    pos_ratio = y_train.mean()
    spw = float((1 - pos_ratio) / (pos_ratio + 1e-12))
    
    models = {
    "Logistic Regression": LogisticRegression(
            solver="liblinear",          
            penalty="l2",               
            C=0.5,                       
            class_weight="balanced",     
            max_iter=2000,
            n_jobs=-1,
            random_state=42
            ),
    
    "KNN Classifier": KNeighborsClassifier(n_neighbors=5,weights="distance"),
   
    "Random Forest": RandomForestClassifier(
            n_estimators=800,
            max_depth=10,
            min_samples_split=4,
            min_samples_leaf=2,
            max_features='sqrt',
            max_samples=0.8,
            class_weight='balanced_subsample',
            n_jobs=-1,
            oob_score=True,
            random_state=42
            ),
    
    #Bonus models
    "XGBoost": XGBClassifier(
            objective='binary:logistic',
            n_estimators=1500,
            learning_rate=0.04,
            max_depth=3, 
            min_child_weight=2,
            subsample=0.7,
            colsample_bytree=0.7,
            reg_lambda=4,
            reg_alpha=0.3,
            eval_metric="auc",
            n_jobs=-1,
            scale_pos_weight=spw,
            random_state = 42
            ),
        
    "MLP Classifier": MLPClassifier(
            hidden_layer_sizes=(256,64),
            activation="relu",
            solver='adam',
            learning_rate_init= 5e-4,
            learning_rate='adaptive',
            alpha=1e-3,
            batch_size=128,
            max_iter=400,
            early_stopping=True,
            validation_fraction=0.2,
            n_iter_no_change=15,
            random_state=42
            )
    }
    


    results = []
    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)[:,1]

        auc = roc_auc_score(y_test,y_prob)
        accuracy = accuracy_score(y_test,y_pred)
        f1score = f1_score(y_test,y_pred)
        balanced_accuracy = balanced_accuracy_score(y_test,y_pred) 
        

        results.append({
            "Model": name,
            "AUC": round(auc, 4),
            "Accuracy": round(accuracy, 4),
            "F1": round(f1score, 4),
            "Balanced Accuracy": round(balanced_accuracy,4)
        })

    results_df = pd.DataFrame(results).sort_values(by="AUC",ascending= False)

    display(results_df)

    return results_df

cls_df = train_classifiers(df, symptom_cols, preprocessor)
```

#### Q3.5 Best Classification Model – Confusion Matrix & Feature Importance (15 points)
