```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

# Ambulance Case Analysis

```python
os.chdir("C:/Users/acer/Desktop/DS701/DS701 data")

df_amb = pd.read_csv("ambulance_cases_in_db.csv")
df_amb = df_amb.drop_duplicates(subset='case_number')
df_amb['status_date'] = pd.to_datetime(df_amb['status_date'])
df_amb['file_date'] = pd.to_datetime(df_amb['file_date'])
df_amb['duration days'] = (df_amb['status_date'] - df_amb['file_date']).dt.days

df_amb_recent_5yr = df_amb[df_amb["file_year"] >= 2020]
df_amb_recent_5yr.head(10)
```

```python
title_prefix = "Ambulance Case filings"

print(f"---{title_prefix} since 2020---")
year_counts = df_amb_recent_5yr["file_year"].value_counts().sort_index()
print(year_counts)

plt.figure(figsize=(10, 6), dpi=300)
plt.plot(
    year_counts,
    marker='o',
    linewidth=2,
    color='green'
)
plt.title(f"{title_prefix} (2020–2025)", fontsize=14, pad=15)
plt.ylabel("Number of Cases")
plt.xlabel("Year")
plt.grid(True, which = 'major', axis = 'both', linestyle ='--', alpha = 0.4)

for x, y in zip(year_counts.index, year_counts.values):
    plt.text(x, y + max(year_counts.values) * 0.04, f"{int(y)}",
             ha="center", va="bottom", fontsize=9)

ymax = max(year_counts.values) * 1.1
plt.ylim(0, ymax)
plt.tight_layout()
plt.show()
```

```python
print("--- ambulance case duration days ---")
df_amb_recent_10yr = df_amb[df_amb["file_year"] >= 2015]
df_amb_recent_10yr = df_amb_recent_10yr.drop_duplicates()
print(df_amb_recent_10yr['duration days'].describe())

plt.figure(figsize=(10, 6), dpi=300)
df_amb_recent_10yr['duration days'].plot(
    kind = 'hist',
    bins = 30
)

plt.title('Ambulance Case Duration Days (2020 - 2025)')
plt.xlabel('Days')
plt.ylabel('Number of Cases')
plt.grid(axis='y', linestyle = '', alpha = 0.5)
plt.tight_layout()
plt.show()
```

```python
top_n = 15
title_prefix = "Ambulance Plaintiffs"

print(f"---Top {top_n} {title_prefix} since 2015---")
top_counts = df_amb_recent_5yr["plaintiff_name"].value_counts().head(top_n)
print(top_counts)

plt.figure(figsize=(10, 6), dpi=300)
top_counts.plot(kind="barh", color="steelblue")
plt.title(f"Top {top_n} {title_prefix} (2020–2025)", fontsize=14, pad=15)
plt.xlabel("Number of Cases")
plt.ylabel("Plaintiff Name")
plt.grid(axis="x", linestyle="--", alpha=0.4)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
import re
def merge_courts(court_name):
    if court_name in ['Central', 'Central Court', 'Boston']:
        return 'Central Court'
    else:
        return court_name

df_amb['court_name_merge'] = df_amb['court_name'].apply(merge_courts)


def extract_case_type(case_num):
    if not isinstance(case_num, str): #if type(case number) != str -> return none
        return none
    match = re.search(r"(SUCV|SU|CV|SC|CR|DV|SP|RE|TR|MH|PM|MI)", case_num.upper())
    return match.group(1) if match else none
    
df_amb['case_type'] = df_amb['case_number'].astype(str).apply(extract_case_type)

type_map = {
    "CV": "Civil",
    "SC": "Small Claims",
    "CR": "Criminal",
    "DV": "Domestic Violence",
    "SP": "Summary Process (Eviction)",
    "RE": "Real Estate",
    "TR": "Traffic",
    "MH": "Mental Health",
    "PM": "Probate",
    "MI": "Miscellaneous",
    "SU": "Superior Court Civil",
    "SUCV": "Superior Court Civil"
}

df_amb['case_type_label'] = df_amb['case_type'].map(type_map).fillna('unknown')
df_amb_recent_5yr = df_amb[df_amb["file_year"] >= 2020]
df_bmc_recent_5yr = df_amb_recent_5yr[df_amb_recent_5yr['court_group'] == 'Boston Municipal Court']


counts = df_bmc_recent_5yr["case_type_label"].value_counts()
main_cats = ["Small Claims", "Civil", "Summary Process (Eviction)"]
counts = counts.rename_axis("Case Type").reset_index(name="Count")
counts["Case Type"] = counts["Case Type"].apply(lambda x: x if x in main_cats else "Others")
counts = (
    counts.groupby("Case Type", as_index=False)["Count"]
    .sum()
    .sort_values("Count", ascending=False)
)
counts = counts.set_index("Case Type")

print("--- Case Type Distribution of Ambulance Cases at Boston Municipal Court(2020-2025) ---")
print(counts["Count"])

plt.figure(figsize= (10,6), dpi = 300)

counts.plot(
    kind='barh', 
    color='steelblue',
    legend = False
)
plt.title("Case Type Distribution of Ambulance Cases at Boston Municipal Court(2020-2025)",fontsize=8)
plt.xlabel('Number of Cases')
plt.grid(axis="x", linestyle="--", alpha=0.4)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

# MSPCA Case Analysis

```python
df_plaintiff = pd.read_csv("refined_medical_cases_with_bmc.csv")

df_plaintiff["file_date"] = pd.to_datetime(df_plaintiff["file_date"], format="%Y%m%d", errors="coerce")
df_plaintiff["file_year"] = df_plaintiff["file_date"].dt.year

keywords = ["animal", "veterinary", "pet"]
df_animals_kw = df_plaintiff[df_plaintiff["keyword"].isin(keywords)].copy()

df_mspca = df_plaintiff[df_plaintiff["plaintiff_name"].str.contains("mspca", case=False, na=False)].copy()
df_mspca_recent_10yr = df_mspca[df_mspca["file_year"] >=2015]

df_animals_combined = pd.concat([df_animals_kw, df_mspca], ignore_index=True)
df_animals_combined = df_animals_combined.drop_duplicates(subset="case_number")

df_animals_combined = df_animals_combined[df_animals_combined["file_year"] >= 2015]

year_counts_animal = (
    df_animals_combined["file_year"]
    .value_counts()
    .sort_index()
)


year_counts_mspca = (
    df_mspca_recent_10yr["file_year"]
    .value_counts()
    .sort_index()
)

plt.figure(figsize=(10, 5), dpi=300)

plt.plot(
    year_counts_animal.index,
    year_counts_animal.values,
    marker= 'o',
    color= 'green',
    linewidth=2,
    label= 'Veterinary/Animal'
)

plt.plot(
    year_counts_mspca.index,
    year_counts_mspca.values,
    marker="o",
    color="blue",
    linewidth=2,
    label= 'MSPCA'
)        

plt.title("Case Filings Comparison (2015–2025)", fontsize=13)
plt.xlabel("Year")
plt.ylabel("Number of Cases")

plt.grid(True, which='major', axis='both', linestyle='--', alpha=0.4)

for x, y in zip(year_counts_animal.index, year_counts_animal.values):
    plt.text(x, y + max(year_counts_animal.values) * 0.02, f"{int(y)}",
             ha="center", va="bottom", fontsize=9)

for x, y in zip(year_counts_mspca.index, year_counts_mspca.values):
    plt.text(x, y - min(year_counts_mspca.values) * 0.45, f"{int(y)}",
             ha="center", va="bottom", fontsize=9)


ymax = max(year_counts_animal.values) * 1.1
ymin = min(year_counts_mspca.values) * 0.5
plt.ylim(ymin, ymax)
plt.legend()
plt.tight_layout()
plt.show()
```
