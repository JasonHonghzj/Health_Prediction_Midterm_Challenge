The values for the species and `other` columns should all be zero or one.

```python
# Create one-hot CSVs for butterflies dataset.
data_dir = "/projectnb/ds542/materials/butterflies"
output_dir = os.path.join(os.getcwd(), "outputs")
os.makedirs(output_dir, exist_ok=True)

train_path = os.path.join(data_dir, "train.csv")
val_path = os.path.join(data_dir, "validation.csv")

train_df = pd.read_csv(train_path)
val_df = pd.read_csv(val_path)


# Ensure expected columns exist
expected_cols = {"uuid", "image_path", "scientific_name"}
missing = expected_cols - set(train_df.columns)
assert not missing, f"Missing columns in train.csv: {missing}"


top9_ordered = train_df["scientific_name"].value_counts().head(9).index.tolist()


def make_onehot(df, top9_names):
    df = df.copy()
    out = df[["uuid", "image_path"]].copy()
    
    # Prepare all zero columns for each top-9 species and 'other'
    for name in top9_names + ["other"]:
        out[name] = 0
    
    # Map rows
    # Use .loc to set the right column to 1
    mask_in_top9 = df["scientific_name"].isin(top9_names)
    

    for name in top9_names:
        out.loc[df["scientific_name"] == name, name] = 1

    out.loc[mask_in_top9 == False, "other"] = 1
    return out

# Build and write train-onehot.csv
train_onehot = make_onehot(train_df, top9_ordered)
train_onehot_path = os.path.join(output_dir, "train-onehot.csv")
train_onehot.to_csv(train_onehot_path, index=False)

val_onehot = make_onehot(val_df, top9_ordered)
val_onehot_path = os.path.join(output_dir, "validation-onehot.csv")
val_onehot.to_csv(val_onehot_path, index=False)

# preview of outputs
result_summary = {
    "train-onehot.csv": train_onehot_path,
    "validation-onehot.csv": val_onehot_path,
    "top9_species_order": top9_ordered
}
result_summary
```

## Part 2: Implement Torch DataSet and DataLoader objects to simplify data access.

[Homework 5](https://colab.research.google.com/github/DL4DS/fa2025/blob/main/static_files/assignments/homework5.ipynb) asked you to implement Torch [DataSet](https://docs.pytorch.org/docs/stable/data.html#torch.utils.data.Dataset) and [DataLoader](https://docs.pytorch.org/docs/stable/data.html#torch.utils.data.DataLoader) objects to separate data loading issues from the training logic.

Implement an appropriate DataSet subclass for the training and validation data.
The class constructor should take in a filename ("train.csv" or "validation.csv") and load the corresponding data.
The `__getitem__` method should return a tuple with a tensor of a single image data and a tensor with the onehot encoding of the target value.

```python
from PIL import Image
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from torchvision.transforms import InterpolationMode


output_dir = os.path.join(os.getcwd(), "outputs")
os.makedirs(output_dir, exist_ok=True)
train_onehot_csv = os.path.join(output_dir, "train-onehot.csv")
val_onehot_csv   = os.path.join(output_dir, "validation-onehot.csv")
data_dir = "/projectnb/ds542/materials/butterflies" 


IMAGENET_MEAN = (0.485, 0.456, 0.406)
IMAGENET_STD  = (0.229, 0.224, 0.225)

class OneHotImageDataset(Dataset):
    def __init__(
        self,
        csv_file: str,
        root_dir: str = None,
        image_col: str = "image_path",
        id_cols: tuple = ("uuid",),
        target_cols: list | None = None,
        img_size: int = 224,
        dtype: torch.dtype = torch.float32,
        transform=None,
        train_mode: bool = True,
    ):
        super().__init__()
        self.df = pd.read_csv(csv_file)
        self.root_dir = root_dir
        self.image_col = image_col
        self.dtype = dtype

        if target_cols is None:
            exclude = {image_col, *id_cols}
            target_cols = [c for c in self.df.columns if c not in exclude]
        self.target_cols = target_cols

        if transform is not None:
            self.transform = transform
        else:
            if train_mode:
                self.transform = transforms.Compose([
                    transforms.RandomResizedCrop(
                        img_size, scale=(0.7, 1.0), ratio=(3/4, 4/3),
                        interpolation=InterpolationMode.BILINEAR
                    ),
                    transforms.RandomHorizontalFlip(p=0.5),
                    transforms.ColorJitter(
                        brightness=0.3, contrast=0.2, saturation=0.15, hue=0.05
                    ),
                    transforms.ToTensor(),
                    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
                ])
            else:
                self.transform = transforms.Compose([
                    transforms.Resize((img_size, img_size), interpolation=InterpolationMode.BILINEAR),
                    transforms.ToTensor(),
                    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
                ])


    def __len__(self):
        return len(self.df)

    def _resolve_path(self, p):
        if self.root_dir is not None and not os.path.isabs(p):
            return os.path.join(self.root_dir, p)
        return p

    def __getitem__(self, idx: int):
        row = self.df.iloc[idx]
        img_path = self._resolve_path(row[self.image_col])
        
        with Image.open(img_path) as im:
            im = im.convert("RGB")
        img_tensor = self.transform(im)

        y = torch.tensor(row[self.target_cols].to_numpy(dtype="float32"), dtype=self.dtype)
        return img_tensor, y
```

Create a DataLoader object for the training data.

```python
# YOUR CHANGES HERE
train_ds = OneHotImageDataset(train_onehot_csv, root_dir=data_dir,train_mode=True)
train_loader = DataLoader(
    train_ds,
    batch_size=64,
    shuffle=True,
    num_workers=4,
    drop_last=True,
    pin_memory=torch.cuda.is_available(),
)

val_ds = OneHotImageDataset(val_onehot_csv, root_dir=data_dir, train_mode=False)
val_loader = DataLoader(
    val_ds,
    batch_size=64,
    shuffle=True,
    num_workers=4,
    drop_last=True,
    pin_memory=torch.cuda.is_available(),
)
```

## Part 3: Define model structure and select an appropriate loss function.

Implement the model class for your convolutional neural network.
The final output of your model should be a tensor of probabilities for each of the ten species choices.

Hint: refer to [homework 5](https://colab.research.google.com/github/DL4DS/fa2025/blob/main/static_files/assignments/homework5.ipynb) and [project 1](https://colab.research.google.com/github/DL4DS/fa2025/blob/main/static_files/assignments/project1.ipynb) for examples.

```python
# YOUR CHANGES HERE
class shallowCNN(torch.nn.Module):
    def __init__(self, num_classes):
        super().__init__()

        self.conv_0 = torch.nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3, padding = 1, device=device)
        self.bn_0 = torch.nn.BatchNorm2d(32)
        
        self.conv_1 = torch.nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding = 1, device=device)
        self.bn_1 = torch.nn.BatchNorm2d(64) 
                
        self.conv_2 = torch.nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3, padding = 1, device=device)
        self.bn_2 = torch.nn.BatchNorm2d(128)
        self.pool_0 = torch.nn.MaxPool2d(2)

        self.conv_3 = torch.nn.Conv2d(in_channels=128, out_channels=256, kernel_size=3, padding = 1, device=device)
        self.bn_3 = torch.nn.BatchNorm2d(256)
        self.pool_1 = torch.nn.MaxPool2d(2)
        
        self.conv_4 = torch.nn.Conv2d(in_channels=256, out_channels=64, kernel_size=3, padding = 1, device=device)
        self.bn_4 = torch.nn.BatchNorm2d(64)

        
        self.flatten = torch.nn.Flatten()
        self.fc_5 = torch.nn.Linear(64*56*56, 256)
        self.fc_6 = torch.nn.Linear(256,num_classes)

        self.relu = torch.nn.ReLU()


    def forward(self, X):

        X = self.relu(self.bn_0(self.conv_0(X)))
        X = self.relu(self.bn_1(self.conv_1(X)))
        X = self.relu(self.pool_0(self.bn_2(self.conv_2(X))))
        X = self.relu(self.pool_1(self.bn_3(self.conv_3(X))))
        X = self.relu(self.bn_4(self.conv_4(X)))
        
        X = self.flatten(X)
        X = self.relu(self.fc_5(X))
        #X = self.dropout1(X)
        X = self.fc_6(X)

        return X

def set_seed(seed=42):
    import random, os
    import numpy as np
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
    random.seed(seed); np.random.seed(seed); torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

set_seed(42)
```

Explain your choices for the shapes in each convolutional layer of your model.

### _YOUR ANSWER HERE_
1. Kernel size = 3, padding = 1 （conv_0~conv_4)  
   A 3×3 kernel with same padding (p=1) is used to preserve spatial dimensions while keeping fine details. Combined with BN and ReLU, it helps stabilize training and reduce the risk of vanishing gradients.  
2. MaxPool2d(2) twice (pool_0, pool_1)  
   Each pooling layer halves the spatial dimensions, systematically reducing noise and computational cost. Based on the input image size of 224×224, two pooling operations reduce it to 56×56, allowing mid- to high-level semantic features to be extracted efficiently.
3. Flatten + Linear (64×56×56 → 256)  
   The 56×56 spatial dimensions is preserved before entering the fully connected layer, allowing the model to leverage local spatial patterns. The feature map is first reduced to 256 dimensions before mapping to the number of classes.
4. BN + ReLU after each convolution  
   Batch Normalization stabilizes the feature distribution and speeds up convergence, while ReLU introduces nonlinearity and sparsity, enabling more stable training even in a relatively shallow network.

Explain your choices for the channels in each convolutional layer of your model.

### _YOUR ANSWER HERE_
1. 3 → 32 → 64 → 128 → 256 (then reduced to 64)
   Referring to common shallow CNN designs, the number of channels doubles in the early layers (32/64/128/256) to increase feature capacity as spatial dimensions shrink through pooling, allowing richer representations to be learned.
2. Bottleneck (conv_4: 256 → 64)
   Before flattening, the channels are reduced to 64 to significantly cut parameters and save memory usage (64×56×56 is much smaller than 256×56×56) while pushing the network to learn more compact and less redundant features, reducing overfitting risk.
3. Channel configuration trade-off
   The 32–256 progression captures features from edges and textures to higher-level semantics, while the final reduction to 64 with a small fully connected head (256 hidden units) strikes a good balance between performance and computational cost.

Pick and define a loss function for your model.
Feel free to use an existing Torch function.

```python
# YOUR CHANGES HERE
model = shallowCNN(num_classes=10).to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=5e-4, weight_decay=5e-4)
loss_function = torch.nn.CrossEntropyLoss()
```

## Part 4: Implement a training loop and explain your choices.

Implement and run an appropriate training loop.
Plot accuracy and loss for the training and validation sets for each epoch.
Plot a summary of the gradients of your choice for each epoch.

```python
def coerce_to_class_indices(targets):
    """
    return to class
    """
    if targets.dim() == 2:
        return targets.argmax(dim=1).long()
    elif targets.dim() == 1:
        if targets.dtype in (torch.int64, torch.int32):
            return targets.long()

        if targets.dtype in (torch.float32, torch.float64):
            return targets.round().long()

    raise ValueError(
        f"targets shape/dtype not supported: shape={tuple(targets.shape)}, dtype={targets.dtype}"
    )
    
def batch_grad_stats(model):
    grads_l2 = []
    for p in model.parameters():
        if p.grad is None:
            continue
        g = p.grad.detach()
        grads_l2.append(g.norm(2))
    if not grads_l2:
        return 0.0, 0.0
    total_l2_norm = torch.norm(torch.stack(grads_l2), p=2).item()
    return total_l2_norm

def reset_cuda():
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        torch.cuda.ipc_collect()
```

```python
# YOUR CHANGES HERE
import math
import gc


scaler = torch.cuda.amp.GradScaler(enabled=torch.cuda.is_available())
num_epochs = 30
print_every = 15

# ES 
early_stop_patience = 3
best_val_loss = float('inf')
epochs_no_improve = 0
best_state = None
best_epoch = 0

best_snapshot = {
    "train_loss": None,
    "train_acc":  None,
    "val_loss":   None,
    "val_acc":    None,
}
best_ckpt_path = os.path.join(output_dir if 'output_dir' in globals() else '.', "best2.pt")


train_losses, val_losses = [], []
train_accs,  val_accs  = [], []
grad_total_l2_per_epoch, grad_mean_abs_per_epoch = [], []

for epoch in range(1, num_epochs + 1):
    # -------- Train phase --------
    model.train()
    train_loss_sum, train_correct, train_count = 0.0, 0, 0
    grad_l2_sum, grad_abs_sum = 0.0, 0.0
    total_train_batches = len(train_loader)

    for b_idx, (inputs, targets) in enumerate(train_loader, 1):
        inputs  = inputs.to(device, non_blocking=True)
        targets = coerce_to_class_indices(targets).to(device)
        
        optimizer.zero_grad()
        
        # AMP forward + loss
        with torch.cuda.amp.autocast(enabled=torch.cuda.is_available()):
            logits = model(inputs)
            loss   = loss_function(logits, targets)
        
        # AMP backward + step
        scaler.scale(loss).backward()
        
        if torch.cuda.is_available():
            scaler.unscale_(optimizer)

        pre_l2 = batch_grad_stats(model)
        
        global_norm = torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=5.0)

        post_l2 = batch_grad_stats(model)
        
        scaler.step(optimizer)
        scaler.update()
        
        batch_l2 = post_l2


        # accumulated train loss / acc
        bs = inputs.size(0)
        train_loss_sum += loss.item() * bs
        preds = logits.argmax(dim=1)
        batch_correct = (preds == targets).sum().item()
        train_correct += batch_correct
        train_count   += bs

        # add each batch's gradients (
        grad_l2_sum   += batch_l2 * bs

        if b_idx % print_every == 0:
            batch_acc = batch_correct / max(1, bs)
            current_print = math.ceil(b_idx // print_every)
            total_prints = math.ceil(total_train_batches // print_every)
            print(f"[Epoch {epoch:02d}][Train][{current_print:02d}/{total_prints}] "
                  f"loss={loss.item():.4f} acc={batch_acc:.4f}")
        
        del logits, loss, preds

    epoch_train_loss = train_loss_sum / max(1, train_count)
    epoch_train_acc  = train_correct  / max(1, train_count)
    train_losses.append(epoch_train_loss)
    train_accs.append(epoch_train_acc)

    # averaged gradient of the epoch
    grad_total_l2_per_epoch.append(grad_l2_sum  / max(1, train_count))
    
    # -------- Validation phase --------
    model.eval()
    val_loss_sum, val_correct, val_count = 0.0, 0, 0
    total_val_batches = len(val_loader)
    
    with torch.inference_mode():
        with torch.no_grad():
            for b_idx, (inputs, targets) in enumerate(val_loader, 1):
                inputs  = inputs.to(device, non_blocking=True)
                targets = coerce_to_class_indices(targets).to(device)
    
                logits = model(inputs)
                loss   = loss_function(logits, targets)
    
                bs = inputs.size(0)
                val_loss_sum += loss.item() * bs
                preds = logits.argmax(dim=1)
                batch_correct = (preds == targets).sum().item()
                val_correct += batch_correct
                val_count   += bs
    
                if b_idx % print_every == 0:
                    batch_acc = batch_correct / max(1, bs)
                    current_print = math.ceil(b_idx // print_every)
                    total_prints = math.ceil(total_val_batches // print_every)
                    print(f"[Epoch {epoch:02d}][Val  ][{current_print:02d}/{total_prints:02d}] "
                          f"loss={loss.item():.4f} acc={batch_acc:.4f}")
    
                del logits, loss, preds

    epoch_val_loss = val_loss_sum / max(1, val_count)
    epoch_val_acc  = val_correct  / max(1, val_count)
    val_losses.append(epoch_val_loss)
    val_accs.append(epoch_val_acc)

    print(
        f"Epoch {epoch:02d}/{num_epochs:02d} | "
        f"train_loss {epoch_train_loss:.4f} , train_acc {epoch_train_acc:.4f} | "
        f"val_loss {epoch_val_loss:.4f} , val_acc {epoch_val_acc:.4f} | "
        f"grad(L2) {grad_total_l2_per_epoch[-1]:.4f}"
    )


    if epoch_val_loss < best_val_loss: 
        best_val_loss = epoch_val_loss
        epochs_no_improve = 0
        best_epoch = epoch
        best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}
        best_snapshot.update({
            "train_loss": epoch_train_loss,
            "train_acc":  epoch_train_acc,
            "val_loss":   epoch_val_loss,
            "val_acc":    epoch_val_acc,
        })
        try:
            torch.save({"state_dict": best_state, "epoch": best_epoch, "metrics": best_snapshot}, best_ckpt_path)
        except Exception as e:
            print(f"[WARN] cannot write in best2.pt：{e}")
    else:
        epochs_no_improve += 1
        if epochs_no_improve >= early_stop_patience:
            print(f"[Early Stopping] go through {early_stop_patience} epoches no improvements on val_loss"
                  f"Early Stoppting Epoch {epoch}. The best Epoch {best_epoch}，val_loss={best_val_loss:.4f}.")
            break
            
    reset_cuda()

if best_state is not None:
    model.load_state_dict(best_state)
    model.eval()
    print(f"[Restore Best] return to best state（Epoch {best_epoch}） | "
          f"train_loss={best_snapshot['train_loss']:.4f}, train_acc={best_snapshot['train_acc']:.4f}, "
          f"val_loss={best_snapshot['val_loss']:.4f}, val_acc={best_snapshot['val_acc']:.4f}")
else:
    print("[Restore Best] no weight snapshot")
```

Explanation

### _YOUR ANSWER HERE_  
1. AMP
I used torch.cuda.amp with GradScaler during training to handle CUDA out-of-memory issues. It helps reduce GPU memory usage and speeds up training, while GradScaler prevents gradient underflow when using mixed precision.

2. Gradient Clipping
I applied global-norm clipping with max_norm=5.0 to all parameters. This keeps gradients from exploding and makes the training curve more stable.

3. Early Stopping (based on val_loss)
I set patience=3 and epochs=30 to avoid overfitting and unnecessary computation. After testing different epoch counts, I found that my model usually reaches its best performance before epoch 30.

4. Gradient Monitoring (batch_grad_stats)
For each batch, I measured the L2-norm of gradients after clipping to check for exploding or vanishing gradients (very large or near zero values).

```python
epochs = range(1,best_epoch + early_stop_patience + 1)

plt.figure(figsize=(6,4), dpi=100)
plt.plot(epochs, train_losses, label='Train Loss')
plt.plot(epochs, val_losses,   label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss per Epoch')
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4), dpi=100)
plt.plot(epochs, train_accs, label='Train Acc')
plt.plot(epochs, val_accs,   label='Val Acc')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy per Epoch')
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4), dpi=100)
plt.plot(epochs, grad_total_l2_per_epoch, label='Total L2 Grad Norm')
plt.xlabel('Epoch')
plt.ylabel('Gradient (units)')
plt.title('Gradient Summary per Epoch')
plt.legend()
plt.tight_layout()
plt.show()
```

## Part 5: Predict species for the test set.

Use your model to predict probabilities for each of the species choices.

```python
# YOUR CHANGES HERE
def eval_transform(img_size: int = 224):
    return transforms.Compose([
        transforms.Resize((img_size, img_size), interpolation=InterpolationMode.BILINEAR),
        transforms.ToTensor(),
        transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
    ])


class TestDataset(Dataset):
    def __init__(
        self :str, 
        csv_file :str,
        root_dir :str, 
        image_col :str ="image_path",
        id_col :str ="uuid",
        img_size :int = 224,
        transform=None
    ):
        self.df = pd.read_csv(csv_file)
        self.root_dir = root_dir
        self.id_col = id_col
        self.image_col = image_col
        self.transform = transform or eval_transform(img_size)

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img_path = os.path.join(self.root_dir, row[self.image_col])
        image = Image.open(img_path).convert("RGB")
        with Image.open(img_path) as im:
            im = im.convert("RGB")
        x = self.transform(im)
        return x, row[self.id_col],row[self.image_col]


test_csv = "/projectnb/ds542/materials/butterflies/test.csv"
inference_transform = eval_transform(img_size=224)
test_ds = TestDataset(
    csv_file=test_csv,
    root_dir=data_dir,
    image_col="image_path",
    id_col="uuid",
    img_size=224,
    transform=inference_transform,
)

test_loader = DataLoader(
    test_ds,
    batch_size=128,
    shuffle=False,
    num_workers=4,
    pin_memory=torch.cuda.is_available(),
)

# Test
model.eval()
probs_list, uuids, paths = [], [], []
with torch.no_grad():
    for imgs, ids, pths in test_loader:
        imgs = imgs.to(device)
        logits = model(imgs)
        probs = torch.softmax(logits, dim=1).cpu()
        probs_list.append(probs)
        uuids.extend(ids)
        paths.extend(pths)

probs_all = torch.cat(probs_list).numpy()
class_names = train_ds.target_cols
```
